```python
import requests
import os
from bs4 import BeautifulSoup
import time
```

```python
SEASONS = list(range(2021,2024))
```

```python
DATA_DIR = "data2"
STANDINGS_DIR =os.path.join(DATA_DIR, "standings")
SCORES_DIR = os.path.join(DATA_DIR,"scores")
```

```python
def get_html(url, selector, sleep=5, retries=3):
    html = None
    for i in range(1, retries+1):
        time.sleep(sleep*i) #stops us from scraping too fast, or we get baned from the server. for each retrie we sleep longer and longer
        try:
            response = requests.get(url) #send a GET request to the URL
            print(response.status_code) #print the status code
            soup = BeautifulSoup(response.text, 'html.parser') #parse the response text with Beautiful Soup
            element = soup.select_one(selector) #locate the element by CSS selector
            html = element.prettify() #get the formatted HTML of the element
        except Exception as e:
            print(f"Error on {url}: {e}")
            continue
        else:
            break
    return html #return the HTML data
```

```python
async def scrape_season(season): 
    url = f"https://www.basketball-reference.com/leagues/NBA_{season}_games.html"
    html = get_html(url, "#content .filter")
    
    soup = BeautifulSoup(html)
    links = soup.find_all("a")
    href = [l["href"] for l in links]
    standings_pages= [f"https://basketball-reference.com{l}" for l in href] #turns them into full urls
    
    for url in standings_pages:
        save_path = os.path.join(STANDINGS_DIR, url.split("/")[-1]) #splits our url based on the forward slash and take the last part, which is essentially the file name 
        if os.path.exists(save_path):
            continue #if we already scraped it dont scrape it again 
            
        html = get_html(url, "#all_schedule")#div element thats just outside the table
        with open(save_path, "w+",encoding="utf-8") as f: #w+ is write mode
            f.write(html)
            
            
```

```python
for season in SEASONS:
    await scrape_season(season) #output should be 200
```

```python
async def scrape_game(standings_file):# will scrape a single standings file and save all of the boxscores

    with open (standings_file, 'r') as f:
        html = f.read()

    soup = BeautifulSoup(html)
    links = soup.find_all("a")
    hrefs = [l.get("href") for l in links]
    box_scores = [l for l in hrefs if l and "boxscore" in l and ".html" in l] #filters out the stuff we dont wanna see , we only want boxscore info
    box_scores = [f"https://www.basketball-reference.com{l}" for l in box_scores] #turns them into urls

    for url in box_scores:
        save_path =  os.path.join(SCORES_DIR, url.split("/")[-1]) #saving it as just the last path of the url
        if os.path.exists(save_path):
            continue #it will ignore teh files it has already scraped

        html = get_html(url, "#content")
        if not html:
            continue 
        with open(save_path, "w+",encoding="utf-8") as f :
            f.write(html)
                    
```

```python
standings_files = [s for s in standings_files if ".html" in s] #filters out any non html files
for f in standings_files :# will loop thru all standings file one by one
        filepath = os.path.join(STANDINGS_DIR,f)
        
        await scrape_game(filepath)
```
