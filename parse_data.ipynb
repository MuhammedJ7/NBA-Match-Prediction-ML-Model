```python
import os 
import pandas as pd
from bs4 import BeautifulSoup 
```

```python
SCORE_DIR = "data2/scores"
```

```python
box_scores = os.listdir(SCORE_DIR)
```

```python
len(box_scores)# amount of boxscores i have 
```

```python
box_scores = [os.path.join(SCORE_DIR, f) for f in box_scores if f.endswith(".html")] # we are linking our boxscores to the full file path
#only files that end with html 
```

```python
def parse_html(box_score):
    with open (box_score, encoding="utf-8") as f:
        html = f.read()
        
    soup = BeautifulSoup(html)
    [s.decompose() for s in soup.select("tr.over_header")]#selecting the table row tags we want to remove
    [s.decompose() for s in soup.select("tr.thead")] #gets rid of reserve line that appears at top of table 
    return soup
```

```python
def create_table(soup):
    # get the date and teams from the h1 element
    h1 = soup.find("h1")
    date, teams = h1.text.split(", ", 1)
    team1, team2 = teams.rsplit(" vs. ", 1) # split the string by " vs. " from the right, only once
    team2 = team2.replace(" Box Score", "") # remove the " Box Score" part from the string

    # get the final score for each team from the table element
    table = soup.find("table", class_="teams")
    rows = table.find_all("tr")
    score1 = rows[0].find("td", class_="right").text # the first row, second column
    score2 = rows[1].find("td", class_="right").text # the second row, second column
    # create a new dataframe with the final score and the team name
    table = pd.DataFrame({"Team": [team1, team2], "Score": [score1, score2]})
    # sort the dataframe by the score in descending order
    table = table.sort_values(by="Score", ascending=False)
    # add the date as the index
    table.index = [date, date]
    # return the table
    return table


box_score= box_scores[0]
soup = parse_html(box_score)
line_score_table=create_table(soup)
```
